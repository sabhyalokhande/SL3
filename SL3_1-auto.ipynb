```python
path="C:/Users/ISHA/Desktop/Datasets/archive/Automobile_data.csv"
df=pd.read_csv(path)
print(df)
```

```python
df["length"] = df["length"].fillna(df["length"].mean())
```

```python
df['normalized-losses']=df['normalized-losses'].replace(to_replace='NaN', value=None)
df['normalized-losses'] = df['normalized-losses'].fillna(value='Missing')
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.replace('?', np.nan, inplace=True)

for col in ['symboling','normalized-losses', 'bore', 'stroke', 'price']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.fillna(df.mean(numeric_only=True), inplace=True)

from sklearn.preprocessing import MinMaxScaler

numeric_cols = ['symboling', 'normalized-losses', 'bore', 'stroke', 'price']
scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

```python
df.columns
```

```python
df['normalized-losses'].value_counts()
```

```python
dummies = pd.get_dummies(df["normalized-losses"])
dummies.head()
```

```python
df = pd.concat([df, dummies], axis=1)
df.drop("normalized-losses", axis = 1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['engine-location'] = le.fit_transform(df['engine-location'])
```
